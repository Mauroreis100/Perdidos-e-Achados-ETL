# Removing inconsistencies on the - Sala / Corredor / WC - column

```python
#Changing complicated instances 
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('D2 - Corredores', 'D2-Corredores')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('Corredores - D1', 'D1-Corredores')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('D1 - WC Rés-do-Chão - Mulher', 'D1-WC Rés-do-Chão - Mulher')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('NE', 'NaN')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('SI3', 'D1-SI3')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('Perto da reprografia', 'D1-Perto da reprografia')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('CEDOC / Biblioteca', 'Biblioteca')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('121', 'SM-121')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('-', 'NaN')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('Antiga 114', 'D2-102')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('102', 'D2-102')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('205', 'E-205')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('106', 'E-102')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('EDIFICIO 1', 'E-Corredores')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('Reprografia', 'D1-Reprografia')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('Biblioteca', 'D1-Biblioteca')
df['Sala / Corredor / WC'] = df['Sala / Corredor / WC'].replace('Biblioteca', 'D1-Biblioteca')
```

# Show every record in - Sala / Corredor / WC - that has '-' sign
And fill in the - Edíficios / Parque de estacionamento / Pontos de Interesse - before the '-' sign

```python
# Find every index of Department and fill department column
for i in range(df['Sala / Corredor / WC'].size):
    if df['Sala / Corredor / WC'][i]!='NaN':
        if df['Sala / Corredor / WC'][i].find('-')>0:
            print(df['Sala / Corredor / WC'][i])
            #print(df['Sala / Corredor / WC'][i][0:df['Sala / Corredor / WC'][i].find("-")],i)
            df['Edíficios / Parque de estacionamento / Pontos de Interesse '][i]=df['Sala / Corredor / WC'][i][0:df['Sala / Corredor / WC'][i].find("-")]
```
